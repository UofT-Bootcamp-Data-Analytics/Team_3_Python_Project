```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statistics import mean
```

```python
inflation_data = pd.read_csv("data/inlation_2015_csv.csv")
inflation_data
```

```python
#calculating the yearly average and creating a new data frame based on year and yearly inflation average
inflation_data = inflation_data.rename(columns={"Measure of core inflation based on a factor model, CPI-common":"Core Inflation"})
date = inflation_data["Month"].tolist()
inflation_rate_monthly = inflation_data["Core Inflation"].tolist()
```

```python
plt.plot(date, inflation_rate_monthly)
plt.title("Inflation Rate per month (2015-current)")
plt.xlabel("Month of year")
plt.xticks(rotation="vertical")
plt.show()
```

```python
#calculating yearly averages from 2015
date_year = []
for i in date:
    aa = i.split('-')
    bb = aa[1]
    date_year.append(bb)
new_df_dic = {"year": date_year, "inflation per month" : inflation_rate_monthly}
new_df = pd.DataFrame(new_df_dic)
yearly_average_inflation = new_df.groupby(["year"])
yearly_average_inflation.first()


inflation_change_perc = []
for index, i in yearly_average_inflation:
    x = i["inflation per month"].tolist()
    inflation_perc = ((x[-1]-x[0])/x[0])*100
    print(inflation_perc)
    inflation_change_perc.append(inflation_perc)
    
        
    
```
