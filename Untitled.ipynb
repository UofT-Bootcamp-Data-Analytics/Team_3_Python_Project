```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import csv

# read data
food_data = './data/food-prices.csv' 
foodPriceData = pd.read_csv(food_data)
food_expenditure_data = './data/Average_food_expenditure_per_household.csv'
foodExpenditureData = pd.read_csv(food_expenditure_data)

food_expense_data = './data/food-expenses.csv' 
foodExpensesData = pd.read_csv(food_expense_data)




# clean foodExpenditureData
foodExpenditureData['Average expenditure per household'] = foodExpenditureData['Average expenditure per household'].str.replace(",", "").str.replace("F", "0").astype(float)
foodExpenditureData['Year'] = foodExpenditureData["Year"].astype(int)

#clean food expendure data
foodExpensesData = foodExpensesData[['REF_DATE', 'Products and product groups', 'VALUE']].copy()
foodExpensesData = foodExpensesData.rename(columns={'REF_DATE': 'Year', 'Products and product groups': 'Products'})
foodExpensesData = foodExpensesData[foodExpensesData['Year'].str.contains('Dec')]
foodExpensesData['Year'] = foodExpensesData['Year'].map(lambda x: x.replace("Dec-", "20"))
foodExpensesData['Year'] = foodExpensesData['Year'].astype(int)

# Get only fish data
salmonExpensesData = foodExpensesData[foodExpensesData['Products'].str.contains(' fish')]
#clean foodPriceData
foodPriceData = foodPriceData.rename(columns={'REF_DATE': 'Year'})
foodPriceData.head()
# salmonExpensesData.head()
```

```python
# Filter by Canada, and get relavent columns
canadaData = foodPriceData[foodPriceData['GEO'] == 'Canada'][['Year', 'Products','VALUE']]
canadaData = canadaData[canadaData["Year"].str.contains("-12")]
canadaData['Year'] = canadaData['Year'].map(lambda x: x.replace("-12", "") )
canadaData['Year'] = canadaData['Year'].astype(int)
canadaData.head()
```

```python
removeProductsFrom = [
    "beef", 
    "pork",
    "chicken",
    "bacon",
    "wieners",
    "shrimp",
    "canned tuna",
    "meatless burgers",
    "milk",
    "cream",
    "butter",
    "margarine",
    "yogurt",
    "eggs",
    "cheese",
    "white bread",
    "flatbread and pita",
    "dry or fresh pasta",
    "brown rice",
    "white rice",
    "cereal",
    "wheat flour",
    "white sugar",
    "apple juice",
    "orange juice",
    "coffee",
    "tea",
    "vegetable oil",
    "canola oil",
    "olive oil",
    "baby food",
    "infant formula",
    "peanut butter",
    "mayonnaise",
    "tofu",
    "hummus",
    "salsa",
    "almonds",
    "peanuts",
    "sunflower seeds",
    "deodorant",
    "toothpaste",
    "shampoo",
    "laundry detergent",
    "dried",
    "dry",
    "sauce",
    "dressing"
]

freshFruits = [ 
    "apples",
    "oranges",
    "bananas",
    "lemons",
    "limes",
    "grapes",
    "cantaloupe",
    "avocado",
    "cucumber",
    "peppers",
    "strawberries",
    "pears" 
]

freshVegs = [
    "sweet potatoes",
    "cabbage",
    "carrots",
    "onions",
    "celery",
    "mushrooms",
    "squash",
    "salad greens",
    "tomatoes",
    "romaine lettuce", 
    "potatoes", 
    "iceberg lettuce",
    "broccoli"
]

others = ["salmon", "canned", "frozen", "ketchup", "crackers", "cookies"]

canadaData['Products'] = canadaData['Products'].str.lower()
filterByProducts = canadaData.copy()
removeProducts = removeProductsFrom + freshFruits + freshVegs
for product in removeProducts: 
    # print(product)
    filterByProducts = filterByProducts[filterByProducts['Products'].str.contains(product) == False]
filterByProducts.head()
```

```python
dataByProduct = filterByProducts.pivot(index='Year', columns='Products', values='VALUE')
dataByProduct.head()
```

```python
dataByProduct.plot()
```

```python
salmonStudy = dataByProduct[['canned salmon, 213 grams', 'salmon, per kilogram']].copy()
salmonStudy['canned salmon, 213 grams'] = salmonStudy['canned salmon, 213 grams'].apply(lambda x: x*1000/213)
salmonStudy = salmonStudy.rename(columns={'canned salmon, 213 grams': 'canned salmon, per kilogram'})
salmonStudy.plot()
```

```python
salmonExpenditureData = foodExpenditureData[foodExpenditureData["Category"].str.contains('Salmon')]
salmonExpenditureData = salmonExpenditureData.pivot(index='Year', columns='Category', values='Average expenditure per household')
salmonExpenditureData.head()
# salmonExpenditureData.plot()
```

```python
# salmonFullStudy = pd.merge(salmonStudy, salmonExpenditureData, on='Year', how='inner')
salmonExpensesData = salmonExpensesData.pivot(index='Year', columns='Products', values='VALUE')


salmonFullStudy = pd.merge(salmonStudy, salmonExpensesData, on='Year', how='inner')


fig, ax = plt.subplots()

ax.plot(salmonFullStudy)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.legend(['Canned Salmon Price', 'Fresh Salmon Price', 'Canned Salmon Expenditure', 'Fresh Salmon Expenditure'],loc='best')
plt.show()
```

```python
# Everything except fruits
removeNotFruits = []
removeNotFruits += removeProductsFrom + others + freshVegs
freshFruitData = canadaData.copy()
for item in removeNotFruits: 
    freshFruitData = freshFruitData[freshFruitData['Products'].str.contains(item) == False]
# freshFruitData.head()

# Everything except veggies
removedNotVegs = []
removedNotVegs += removeProductsFrom + others + freshFruits
freshVegData = canadaData.copy()
for item in removedNotVegs: 
    freshVegData = freshVegData[freshVegData['Products'].str.contains(item) == False]
# freshVegData.head()

# Everything except canned
removeNotCanned = others.copy()
removeNotCanned.remove("canned")
removeNotCanned += removeProductsFrom + freshFruits + freshVegs + removeNotCanned
cannedFoodData = canadaData.copy()
for item in removeNotCanned: 
    cannedFoodData = cannedFoodData[cannedFoodData['Products'].str.contains(item) == False]
# cannedFoodData.head(25)

# Everything except frozen
removeNotFrozen = others.copy()
removeNotFrozen.remove("frozen")
removeNotFrozen += removeProductsFrom + freshFruits + freshVegs + removeNotFrozen
frozenFoodData = canadaData.copy()
for item in removeNotFrozen: 
    frozenFoodData = frozenFoodData[frozenFoodData['Products'].str.contains(item) == False]
# frozenFoodData.head()

cannedFoodData.head(25)
```

```python
def consolidateDatas(data, name): 
    dataset = data.copy()
    dataset["summed price"] = dataset.groupby(["Year"])["VALUE"].transform(sum)
    dataset = dataset.drop(['Products', 'VALUE'], axis=1)
    dataset = dataset.drop_duplicates()
    dataset = dataset.rename(columns={"summed price": "VALUE"})
    dataset["Products"] = name
    return dataset
    
consolidatedFreshFruits = consolidateDatas(freshFruitData, "Fresh fruit")
consolidatedFreshVegs = consolidateDatas(freshVegData, "Fresh vegetables")
consolidatedCannedFoods = consolidateDatas(cannedFoodData, "")
consolidatedFrozenFoods = consolidateDatas(frozenFoodData)

# consolidatedFreshFruits.head()
# consolidatedFreshVegs.head()
# consolidatedCannedFoods.head()
consolidatedFrozenFoods.head()
```
